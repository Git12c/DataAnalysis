```python
df3_mod= df3[['ProductID','OrderDate', 'ShipBranch', 'ShipDate', 'TotalQTY', 'ExtTotalSales']]
df3_mod= df3_mod.rename(columns={'OrderDate':'Order Date', 'ShipBranch':'Ship Branch', 'ShipDate':'Ship Date', 'TotalQTY':'Total QTY', 'ExtTotalSales':'Ext Total Sales'})
df3_mod.head()
```

```python
df3_num_products= df3_mod['ProductID'].unique()
df3_num_branches= df3_mod['Ship Branch'].unique()
df3_num_branches, df3_num_products
```

```python
df_merged= pd.concat([df1_mod, df2_mod, df3_mod])
df_merged
```

```python
df_merged['Order Date']= pd.to_datetime(df_merged['Order Date'], format= 'mixed', dayfirst=True)

df_merged['Ship Date']= pd.to_datetime(df_merged['Ship Date'], format= 'mixed', dayfirst=True)
```

```python
unique_products= df_merged['ProductID'].unique()

unique_products
```

```python
target_productID= 500000

targeted_data= df_merged[df_merged['ProductID']== target_productID]

targeted_data
```

```python
target_data_prophet= targeted_data.rename(columns={'Ship Date':'ds', 'Ext Total Sales':'y'})
target_data_prophet['ds']= pd.to_datetime(target_data_prophet['ds'])
```

```python
target_data_prophet
```

```python
target_data_prophet['y'] = (
    target_data_prophet['y']
    .replace('[\$,]', '', regex=True)              # Remove $ and ,
    .replace('[()]', '', regex=True)               # Remove parentheses
    .astype(str)
)

# Detect and handle negatives (originally in parentheses)
target_data_prophet['y'] = target_data_prophet['y'].apply(lambda x: -float(x) if '(' in x or ')' in x else float(x))
```

```python
target_data_prophet['Ship Branch'].unique()
branch_1_data = target_data_prophet[target_data_prophet['Ship Branch'] == 'Branch 1']
branch_1_data.head()
```

```python
target_data_prophet_ds= branch_1_data[['ds','y']]
target_data_prophet_ds
```

```python
from prophet import Prophet

model= Prophet()
model.fit(target_data_prophet_ds)
```

```python
future= pd.DataFrame()

num_future_days= 90

future= model.make_future_dataframe(periods=num_future_days, freq='D')

future.head()
```

```python
forecast= pd.DataFrame()

forecast= model.predict(future)

forecast.head()
```

```python
fig1= model.plot(forecast, uncertainty=True, plot_cap=True, include_legend=True)
plt.xlabel('Date')
plt.ylabel('Total price')
plt.show()
```

```python
fig2= model.plot_components(forecast)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# --- Custom start and end dates ---
plot_start_date = pd.to_datetime("2016-02-29")  # Replace as needed
plot_end_date = pd.to_datetime("2242-01-11")    # Replace as needed

# --- Sanity check for date range ---
if plot_end_date > forecast['ds'].max():
    print("Plot end date is beyond forecast range.")
else:
    # --- Slice forecast within custom range ---
    plot_data = forecast[(forecast['ds'] >= plot_start_date) & (forecast['ds'] <= plot_end_date)].copy()

    # --- Slice actual/historical data (assuming it's in target_data_prophet_ds) ---
    historical_plot_data = target_data_prophet_ds[
        (target_data_prophet_ds['ds'] >= plot_start_date) & 
        (target_data_prophet_ds['ds'] <= plot_end_date)
    ].copy()

    if not plot_data.empty:
        plt.figure(figsize=(12, 6))

        # Plot historical sales
        if not historical_plot_data.empty:
            plt.plot(historical_plot_data['ds'], historical_plot_data['y'],
                     label='Actual Sales', color='blue', marker='.', linestyle='-')

        # Plot forecast mean
        plt.plot(plot_data['ds'], plot_data['yhat'], label='Forecast (Mean)', color='red', linestyle='-')

        # Confidence interval
        plt.fill_between(plot_data['ds'], plot_data['yhat_lower'], plot_data['yhat_upper'],
                         color='red', alpha=0.2, label='Confidence Interval')

        # Decorations
        plt.title(f'Actual & Forecast Sales: {plot_start_date.date()} to {plot_end_date.date()}')
        plt.xlabel('Date')
        plt.ylabel('Sales')
        plt.legend()
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print("No forecast data in selected date range.")
```
